```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

path = '../images/board.jpg'
image = cv.imread(path)
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#Create kernel
kernel = np.ones((3, 3), np.float32)/9 #blur kernel
kernel2 = np.array([[0, -1, 0], #sharp kernel
                   [-1, 5, -1],
                   [0, -1, 0]])
print(kernel)
print(kernel2)

#Convolved image
convolved_blur = cv.filter2D(image,-1,kernel)
#convolved_blur = cv.blur(image, (5,5))
convolved_sharp = cv.filter2D(image,-1,kernel2)

#Plot image
fig = plt.figure(figsize=(25, 10))
fig.tight_layout(pad=3)
#Ax1
ax1 = fig.add_subplot(1, 3, 1)
ax1.set_title("Original image")
ax1.imshow(image)
#Ax2
ax2 = fig.add_subplot(1, 3, 2)
ax2.set_title("Convolution Blur Kernel")
ax2.imshow(convolved_blur)
#Ax3
ax3 = fig.add_subplot(1, 3, 3)
ax3.set_title("Convolution Sharp Kernel")
ax3.imshow(convolved_sharp)
```
