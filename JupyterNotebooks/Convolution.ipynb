```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

path = '../images/board.jpg'
image = cv.imread(path)
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)

fig, ax = plt.subplots(1, figsize=(12,8))
plt.imshow(image)

kernel = np.ones((3, 3), np.float32)/9
kernel2 = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])
print(kernel)
print(kernel2)

#convolved_blur = cv.filter2D(image,-1,kernel)
convolved_blur = cv.blur(image, (5,5))
fig, ax = plt.subplots(1, figsize=(12,8))
plt.imshow(convolved_blur)

convolved_sharp = cv.filter2D(image,-1,kernel2)
fig, ax = plt.subplots(1, figsize=(12,8))
plt.imshow(convolved_sharp)
```
