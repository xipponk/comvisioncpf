```python
import cv2
import matplotlib.pyplot as plt


img = cv2.imread('BasicImage.png')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
print_im_value(img)
plt.axis('off')
```

#### Erosion

```python
kernel = np.ones((3,3),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
plt.imshow(erosion)
plt.axis('off')
print_im_value(erosion)
```

#### Dilation

```python
kernel = np.ones((3,3),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 1)
plt.imshow(dilation)
plt.axis('off')
print_im_value(dilation)
```

#### Opening 
erosion -> dilation

```python
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
```

#### Closing
dilation -> erosion

```python
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
```

#### Morphological Gradient
difference between dilation and erosion

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
```

```python
# Create a kernel that we will use to sharpen our image
# an approximation of second derivative, a quite strong kernel
kernel = np.array([[1, 1, 1], [1, -8, 1], [1, 1, 1]], dtype=np.float32)
# do the laplacian filtering as it is
# well, we need to convert everything in something more deeper then CV_8U
# because the kernel has some negative values,
# and we can expect in general to have a Laplacian image with negative values
# BUT a 8bits unsigned int (the one we are working with) can contain values from 0 to 255
# so the possible negative number will be truncated
imgLaplacian = cv2.filter2D(egg_RGB, cv2.CV_32F, kernel)
sharp = np.float32(egg_RGB)
imgResult = sharp - imgLaplacian
# convert back to 8bits gray scale
imgResult = np.clip(imgResult, 0, 255)
imgResult = imgResult.astype('uint8')
imgLaplacian = np.clip(imgLaplacian, 0, 255)
imgLaplacian = np.uint8(imgLaplacian)
#cv.imshow('Laplace Filtered Image', imgLaplacian)
plt.imshow(imgResult)
```

```python
def filter_contours_img(contours, img_draw, color_bbox):
    count = 0
    for c in contours:
        rect = cv2.boundingRect(c)
        x,y,w,h = rect
        area = w * h
        if area > 1000:
            count = count + 1 # นับ object ที่มีพื้นที่มากกว่า 1000 pixel
            cv2.rectangle(img_draw, (x, y), (x+w, y+h), color_bbox, 5)
    return img_draw, count
```

```python
def color_segmentation(im_BGR,target_HSV=[28,0.6,166],threshold=200):
  im_HSV=cv2.cvtColor(im_BGR.astype(np.float32),cv2.COLOR_BGR2HSV_FULL)
  segment=255-np.linalg.norm(im_HSV-target_HSV,axis=2)
  segment=(segment>threshold).astype(np.uint8)
  kernel = np.ones((5,5),np.uint8)
  segment = cv2.morphologyEx(segment, cv2.MORPH_OPEN, kernel)
  segment = cv2.morphologyEx(segment, cv2.MORPH_CLOSE, kernel)
  return segment
```

```python
def find_ellipse(im0,raw_im=None,area_thres=40):
  conts=[]
  im=cv2.distanceTransform(im0,cv2.DIST_L2, cv2.DIST_MASK_PRECISE)>7
  im=im.astype(np.uint8)
  contours = cv2.findContours(im,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
  if raw_im is None:
    raw_im=im;
  for  c in contours[0]:
    if cv2.contourArea(c)>area_thres and len(c)>5:
      ellipse = cv2.fitEllipse(c)
      cv2.ellipse(raw_im,ellipse,(255,0,0),2)
      conts.append(ellipse)
  return raw_im
```
