```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

path = r'../images/trex.png'
# Load image in RGB scale
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Open image
fig, ax = plt.subplots(1, figsize=(12,8))
ax.set(title='Original')
plt.imshow(image)
print("max of 255: {}".format(cv2.add(np.uint8([200]), np.uint8([100]))))
print("max of 0: {}".format(cv2.subtract(np.uint8([50]), np.uint8([100]))))
print("wrap around: {}".format(np.uint8([200]) + np.uint8([100])))
print("wrap around: {}".format(np.uint8([50]) - np.uint8([100])))

M = np.ones(image.shape, dtype = "uint8") * 100
added = cv2.add(image, M)
fig, ax = plt.subplots(1, figsize=(12,8))
ax.set(title='Added')
plt.imshow(added)

M = np.ones(image.shape, dtype = "uint8") * 100
subtracted = cv2.subtract(image, M)
fig, ax = plt.subplots(1, figsize=(12,8))
ax.set(title='Subtracted')
plt.imshow(subtracted)

'''
hist_red = cv2.calcHist([image], [2], None, [256], [0,256])
hist_green = cv2.calcHist([image], [1], None, [256], [0,256])
hist_blue = cv2.calcHist([image], [0], None, [256], [0,256])
fig, ax = plt.subplots()
ax.plot(hist_red, color='r', label='RED')
ax.plot(hist_green, color='g', label='GREEN')
ax.plot(hist_blue, color='b', label='BLUE')
ax.set_title("Original")
ax.legend()

hist_red = cv2.calcHist([added], [2], None, [256], [0,256])
hist_green = cv2.calcHist([added], [1], None, [256], [0,256])
hist_blue = cv2.calcHist([added], [0], None, [256], [0,256])
fig, bx = plt.subplots()
bx.plot(hist_red, color='r', label='RED')
bx.plot(hist_green, color='g', label='GREEN')
bx.plot(hist_blue, color='b', label='BLUE')
bx.set_title("Added")
bx.legend()

hist_red = cv2.calcHist([subtracted], [2], None, [256], [0,256])
hist_green = cv2.calcHist([subtracted], [1], None, [256], [0,256])
hist_blue = cv2.calcHist([subtracted], [0], None, [256], [0,256])
fig, cx = plt.subplots()
cx.plot(hist_red, color='r', label='RED')
cx.plot(hist_green, color='g', label='GREEN')
cx.plot(hist_blue, color='b', label='BLUE')
cx.set_title("Subtracted")
cx.legend()
'''
```
