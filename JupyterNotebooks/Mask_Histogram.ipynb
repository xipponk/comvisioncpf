```python
from matplotlib import pyplot as plt
from imutils import plot_histogram
import numpy as np
import cv2

path = r'../images/beach.png'
image = cv2.imread(path)

cv2.startWindowThread()
cv2.imshow("Original Image", image)
plot_histogram(image, "Histogram of Original Image")

mask = np.zeros(image.shape[0:2], dtype='uint8')
cv2.rectangle(mask, (15,15), (130, 100), 255, -1)
cv2.imshow("Mask", mask)

masked = cv2.bitwise_and(image, image, mask = mask)
cv2.imshow("Applying the Mask", masked)
plot_histogram(image, "Histogram of Masked Image", mask=mask)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
