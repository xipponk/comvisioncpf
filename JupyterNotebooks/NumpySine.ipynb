```python
import numpy as np
import matplotlib.pyplot as plt

N = 256
A = 127.0
W = 8.0
P = 45.0
x = np.linspace(-np.pi,np.pi, N)
sine1D = P + (A * np.sin(x * W))
sine1D = np.uint8(sine1D)
sine2D = np.ndarray((N,N), dtype=np.uint8)

for i in range(N):
    sine2D[i]= np.roll(sine1D,-i)  # shift the 1D sin data by -i, -i increases with rows

#Plot image
fig = plt.figure(figsize=(5, 5))
ax1 = fig.add_subplot(1,1,1)
ax1.set(title='2D Sine wave')
ax1.imshow(sine2D, cmap='gray')
```
