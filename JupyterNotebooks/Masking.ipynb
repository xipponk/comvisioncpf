```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import imutils

path = r'../images/beach.png'
Original = cv2.imread(path)
Original = cv2.cvtColor(Original, cv2.COLOR_BGR2RGB)

#Create canvas and mask
Rectangle = np.zeros(Original.shape[0:2], dtype="uint8")
Circle = np.zeros(Original.shape[0:2], dtype="uint8")
(cX, cY) = (Original.shape[1] // 2, Original.shape[0] // 2) #find coordinate of center point
cv2.rectangle(Rectangle, (cX-75, cY-75), (cX+75, cY+75), 255, -1) #Create rectangle mask 150x150 on canvas
cv2.circle(Circle, (cX, cY), 100, 255, -1) #Create circle mask with radius 100 on canvas

#Masking to image
Rectangle_masked = cv2.bitwise_and(Original, Original, mask=Rectangle)
Circle_masked = cv2.bitwise_and(Original, Original, mask=Circle)

#Plot image to Matplotlib
fig = plt.figure(figsize=(20, 10)) #create figure
fig.tight_layout(pad=3)
#Ax1
ax1 = fig.add_subplot(2, 3, 1)
ax1.set_title("Original Image")
ax1.set_xlabel('X Pixels')
ax1.set_ylabel('Y Pixels')
ax1.imshow(Original)
#Ax2
ax1 = fig.add_subplot(2, 3, 2)
ax1.set_title("Rectangle")
ax1.set_xlabel('X Pixels')
ax1.set_ylabel('Y Pixels')
ax1.imshow(Rectangle, cmap='Pastel1')
#Ax3
ax1 = fig.add_subplot(2, 3, 3)
ax1.set_title("Circle")
ax1.set_xlabel('X Pixels')
ax1.set_ylabel('Y Pixels')
ax1.imshow(Circle, cmap='Pastel1')
#Ax4
ax1 = fig.add_subplot(2, 3, 4)
ax1.set_title("Original Image")
ax1.set_xlabel('X Pixels')
ax1.set_ylabel('Y Pixels')
ax1.imshow(Rectangle_masked)
#Ax5
ax1 = fig.add_subplot(2, 3, 5)
ax1.set_title("Original Image")
ax1.set_xlabel('X Pixels')
ax1.set_ylabel('Y Pixels')
ax1.imshow(Circle_masked)
```
