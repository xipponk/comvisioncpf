```python
import numpy as np
from matplotlib import pyplot as plt

N = 30
xf = np.zeros((N,N), float)
xf[0, 5] = 1.0
xf[0, N-5] = 1.0
ixf = np.fft.ifft2(xf)

yf = np.zeros((N,N), float)
yf[5, 0] = 1.0
yf[N-5, 0] = 1.0
iyf = np.fft.ifft2(yf)

uv = np.zeros((N,N), float)
uv[5, 10] = 1.0
uv[N-5, N-10] = 1.0
iuv = np.fft.ifft2(uv)

#Plot image
fig = plt.figure(figsize=(7, 12))
#######################################
ax1 = fig.add_subplot(3,2,1)
ax1.set(title='Frequency domain')
#ax1.spines['left'].set_position('center')
#ax1.spines['bottom'].set_position('center')
#ax1.spines['right'].set_color('none')
#ax1.spines['top'].set_color('none')
#ax1.xaxis.set_ticks_position('bottom')
#ax1.yaxis.set_ticks_position('left')
ax1.imshow(xf, interpolation='nearest', origin='lower', cmap='Reds')
#######################################
ax2 = fig.add_subplot(3,2,2)
ax2.set(title='2D Sine wave in time domain')
ax2.imshow(np.real(ixf), interpolation='bicubic', origin='lower', cmap='gray')
#######################################
ax3 = fig.add_subplot(3,2,3)
ax3.set(title='Frequency domain')
ax3.imshow(yf, interpolation='nearest', origin='lower', cmap='Reds')
#######################################
ax4 = fig.add_subplot(3,2,4)
ax4.set(title='2D Sine wave in time domain')
ax4.imshow(np.real(iyf), interpolation='bicubic', origin='lower', cmap='gray')
#######################################
ax5 = fig.add_subplot(3,2,5)
ax5.set(title='Frequency domain')
ax5.imshow(uv, interpolation='nearest', origin='lower', cmap='Reds')
#######################################
ax6 = fig.add_subplot(3,2,6)
ax6.set(title='2D Sine wave in time domain')
ax6.imshow(np.real(iuv), interpolation='bicubic', origin='lower', cmap='gray')
```
