```python
from matplotlib import pyplot as plt
import numpy as np
import cv2

path = r'../images/beach.png'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.cvtColor(image, cv2.COLOR_2GRAY)

#Equalized image
eq = cv2.equalizeHist(image)

#Calculate histogram
original_hist = cv2.calcHist([image], [0], None, [256], [0, 256])
eq_hist = cv2.calcHist([eq], [0], None, [256], [0, 256])

#Plot image
fig1 = plt.figure(figsize=(20, 5))
ax1 = fig1.add_subplot(1,2,1)
ax1.set_title('Original Grray Image')
ax1.imshow(image, cmap='gray')
ax2 = fig1.add_subplot(1,2,2)
ax2.set_title('Equalized Image')
ax2.imshow(eq, cmap='gray')

#Plot histogram
fig2 = plt.figure(figsize=(20, 5))
ax1 = fig2.add_subplot(1,2,1)
ax1.set_title('Original Histogram')
ax1.set_xlabel('Bins')
ax1.set_xlim([0, 255])
ax1.set_ylabel('# of pixels')
ax1.plot(original_hist, color='k')
ax1 = fig2.add_subplot(1,2,2)
ax1.set_title('Equalized Histogram')
ax1.set_xlabel('Bins')
ax1.set_xlim([0, 255])
ax1.set_ylabel('# of pixels')
ax1.plot(eq_hist, color='k')
```
