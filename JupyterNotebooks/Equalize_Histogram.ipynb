```python
from matplotlib import pyplot as plt
import numpy as np
import cv2

path = r'../images/beach.png'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2.startWindowThread()

fig, ax = plt.subplots(figsize=(10, 5))
ax.set(title='Original Histogram', xlabel='Bins', ylabel='# of Pixels', xlim=[0, 256]);
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
ax.plot(hist, color='k')
plt.show()

eq = cv2.equalizeHist(image)
cv2.imshow("Equalized Image", np.hstack([image, eq]))

fig, ax = plt.subplots(figsize=(10, 5))
ax.set(title='Equalized Histogram', xlabel='Bins', ylabel='# of Pixels', xlim=[0, 256]);
hist = cv2.calcHist([eq], [0], None, [256], [0, 256])
ax.plot(hist, color='k')
plt.show()



cv2.waitKey(0)
cv2.destroyAllWindows()
```
