```python
from matplotlib import pyplot as plt
import numpy as np
import cv2

#Opean and load image
path = r'images/coins.png'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#Blur image
#image = cv2.GaussianBlur(image, (5,5), 1.3, cv2.BORDER_REPLICATE) #Best value occur at sigma of guassian = 1.3
image = cv2.GaussianBlur(image, (5,5), 1.3)

#Canny edge detection with suggest parameter
v = np.median(image)
sigma = 0.33
lower = int((1.0 - sigma)*v)
upper = int((1.0 + sigma)*v)
edged = cv2.Canny(image, lower, upper)

#Plot the first method
fig = plt.figure(figsize=(15, 9)) #create figure
fig.tight_layout(pad=3)
#Ax1
ax1 = fig.add_subplot(1, 2, 1)
ax1.set_title("Original image")
ax1.imshow(image, cmap='gray')
#Ax2
ax2 = fig.add_subplot(1, 2, 2)
ax2.set_title("Canny Edge Detection (Auto)")
ax2.imshow(edged, cmap='gray')
```
