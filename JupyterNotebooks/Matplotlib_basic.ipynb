```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

imageBGR = cv.imread('../images/messi.jpg')
imageRGB = cv.cvtColor(imageBGR, cv.COLOR_BGR2RGB)

#Open image with OpenCV
cv.startWindowThread()
cv.imshow("BGR", imageBGR)
cv.imshow("RGB", imageRGB)
cv.waitKey(0)
cv.destroyAllWindows()

#Open image with Matplotlib
plt.figure()
plt.imshow(imageRGB)
plt.title('RGB image')
plt.xlabel('width')
plt.ylabel('height')
plt.show()
'''
plt.imshow(imageBGR)
plt.title('RGB image')
plt.xlabel('width')
plt.ylabel('height')
'''
plt.figure()
plt.subplot(121),plt.imshow(imageRGB),plt.title('RGB image'),plt.xlabel('width'),plt.ylabel('height')
plt.subplot(122),plt.imshow(imageBGR),plt.title('BGR image'),plt.xlabel('width'),plt.ylabel('height')
plt.show()

#Object based plotting
fig1 = plt.figure() #create figure object
fig1.suptitle('RGB Image')
ax1 = fig1.add_subplot(1,1,1) #create an axes object in the figure
ax1.set(title='Messi', xlabel='width', ylabel='height')
ax1.imshow(imageRGB)

fig2 = plt.figure(figsize=(10, 3))
fig2.suptitle('Open image with Matplotlib')
fig2.tight_layout(pad=1.08)
ax1 = fig2.add_subplot(1,2,1)
ax2 = fig2.add_subplot(1,2,2)
ax1.set(title='RGB Image', xlabel='width', ylabel='height')
ax2.set(title='BGR Image', xlabel='width', ylabel='height')
ax1.imshow(imageRGB)
ax2.imshow(imageBGR)

#Unpacking methods
fig3, ax = plt.subplots(1, 2, figsize=(10, 4))
fig3.suptitle('Plot with object method')
ax[0].set(title='RGB Image', xlabel='width', ylabel='height')
ax[1].set(title='BGR Image', xlabel='width', ylabel='height')
ax[0].imshow(imageRGB)
ax[1].imshow(imageBGR)
```
