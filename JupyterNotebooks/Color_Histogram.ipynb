```python
from matplotlib import pyplot as plt
import numpy as np
import cv2

path = r'../images/trex.png'
original = cv2.imread(path)
original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)

chans = cv2.split(original)
colors = ('b', 'g', 'r')
histB = cv2.calcHist([chans[0]], [0], None, [256], [0, 256])
histG = cv2.calcHist([chans[1]], [0], None, [256], [0, 256])
histR = cv2.calcHist([chans[2]], [0], None, [256], [0, 256])

fig = plt.figure(figsize=(20, 5))
ax1 = fig.add_subplot(1,3,1)
ax1.set_title('Red Histogram')
ax1.set_xlabel('Bins')
ax1.set_xlim([0, 255])
ax1.set_ylabel('# of pixels')
ax1.plot(histR, color='r')

ax2 = fig.add_subplot(1,3,2)
ax2.set_title('Green Histogram')
ax2.set_xlabel('Bins')
ax2.set_xlim([0, 255])
ax2.set_ylabel('# of pixels')
ax2.plot(histG, color='g')

ax3 = fig.add_subplot(1,3,3)
ax3.set_title('Blue Histogram')
ax3.set_xlabel('Bins')
ax3.set_xlim([0, 255])
ax3.set_ylabel('# of pixels')
ax3.plot(histB, color='b')

fig1 = plt.figure(figsize=(7, 5))
ax1 = fig1.add_subplot(1,1,1)
ax1.set_title('Color Histogram')
ax1.set_xlabel('Bins')
ax1.set_xlim([0, 255])
ax1.set_ylabel('# of pixels')
ax1.plot(histR, color='r')
ax1.plot(histG, color='g')
ax1.plot(histB, color='b')
```
