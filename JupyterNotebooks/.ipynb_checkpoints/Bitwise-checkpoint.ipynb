```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

canvas1 = np.zeros((300, 300), dtype="uint8")
cv2.rectangle(canvas1, (25,25), (275,275), 255, -1)
fig, ax = plt.subplots(1)
ax.set(title='Rectangle')
plt.imshow(canvas1, cmap='gray')

canvas2 = np.zeros((300, 300), dtype="uint8")
cv2.circle(canvas2, (150,150), 150, 255, -1)
fig, ax = plt.subplots(1)
ax.set(title='Circle')
plt.imshow(canvas2, cmap='gray')

bitwiseAND = cv2.bitwise_and(canvas1, canvas2)
fig, ax = plt.subplots(1)
ax.set(title='AND')
plt.imshow(bitwiseAND, cmap='gray')

bitwiseOR = cv2.bitwise_or(canvas1, canvas2)
fig, ax = plt.subplots(1)
ax.set(title='OR')
plt.imshow(bitwiseOR, cmap='gray')

bitwiseXOR = cv2.bitwise_xor(canvas1, canvas2)
fig, ax = plt.subplots(1)
ax.set(title='XOR')
plt.imshow(bitwiseXOR, cmap='gray')

bitwiseNOT = cv2.bitwise_not(canvas1)
fig, ax = plt.subplots(1)
ax.set(title='NOT Rectangle')
plt.imshow(bitwiseNOT, cmap='gray')

bitwiseNOT = cv2.bitwise_not(canvas2)
fig, ax = plt.subplots(1)
ax.set(title='NOT Circle')
plt.imshow(bitwiseNOT, cmap='gray')
```
