```python
from matplotlib import pyplot as plt
import numpy as np
import cv2

path = r'../images/beach.png'
image = cv2.imread(path)

cv2.startWindowThread()
cv2.imshow("Original", image)
chans = cv2.split(image)

fig, ax = plt.subplots(3, 1, figsize=(5,10))

ax[0].set(title='2D Color Histogram for G and B');
hist = cv2.calcHist([chans[1], chans[0]], [0, 1], None, [32, 32], [0, 256, 0, 256])
p = ax[0].imshow(hist, interpolation='nearest')
plt.colorbar(p, ax=ax[0])

ax[1].set(title='2D Color Histogram for G and R');
hist = cv2.calcHist([chans[1], chans[2]], [0, 1], None, [32, 32], [0, 256, 0, 256])
p = ax[1].imshow(hist, interpolation='nearest')
plt.colorbar(p, ax=ax[1])

ax[2].set(title='2D Color Histogram for B and R');
hist = cv2.calcHist([chans[0], chans[2]], [0, 1], None, [32, 32], [0, 256, 0, 256])
p = ax[2].imshow(hist, interpolation='nearest')
plt.colorbar(p, ax=ax[2])

print('2D Histogram shape: {}, with {} values'.format(hist.shape, hist.flatten().shape[0]))

cv2.waitKey(0) 
cv2.destroyAllWindows()
```
