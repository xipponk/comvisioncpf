```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread('../images/messi.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

fig, ax = plt.subplots()
ax.set(title="Read image")
ax.imshow(img)

#Access image properties
print('The image height {}, width {}, and {} channels'.format(img.shape[0], img.shape[1], img.shape[2]))
```
