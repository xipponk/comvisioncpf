```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

image = cv.imread('../images/messi.jpg')
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)

fig, ax = plt.subplots()
ax.set(title="Read image")
ax.imshow(image)

#Access image properties, image size and image datatype
print('The image has height {} and width {} with {} channels.'.format(image.shape[0], image.shape[1], image.shape[2]))
print('The size of image is {} pixels.'.format(image.size))
print('Datatype of image is {}.'.format(image.dtype))

#Image Region of Interest (ROI)
roi = image.copy()
ball = image[280:340, 330:390]
roi[273:333, 100:160] = ball

fig2, ax = plt.subplots()
ax.set(title="ROI technic")
ax.imshow(roi)

#Split and merge image
#r, g, b = cv.split(image)
r = image[:,:,0]
g = image[:,:,1]
b = image[:,:,2]
fig3, ax = plt.subplots(1, 3, figsize=(30, 10))
ax[0].set(title='Red channel')
ax[0].imshow(r, cmap='gray')
ax[1].set(title='Green channel')
ax[1].imshow(g, cmap='gray')
ax[2].set(title='Blue channel')
ax[2].imshow(b, cmap='gray')
merged = cv.merge((r,g,b))
fig4, ax = plt.subplots()
ax.set(title='Merged image')
ax.imshow(merged)

#filled Red channel with zero
image2 = image.copy()
image2[:,:,0] = 0
R = image2[:,:,0]
G = image2[:,:,1]
B = image2[:,:,2]
merged2 = cv.merge((R,G,B))
fig5, ax = plt.subplots()
ax.set(title='Image with "0" Red channel')
ax.imshow(merged2)
```
