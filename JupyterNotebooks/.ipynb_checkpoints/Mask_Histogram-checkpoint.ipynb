```python
from matplotlib import pyplot as plt
from imutils import plot_histogram
import numpy as np
import cv2

path = r'../images/beach.png'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#calculte histogram
chans = cv2.split(image)
histR = cv2.calcHist([chans[0]], [0], None, [256], [0, 256])
histG = cv2.calcHist([chans[1]], [0], None, [256], [0, 256])
histB = cv2.calcHist([chans[2]], [0], None, [256], [0, 256])

#plot image and histogram
fig1 = plt.figure(figsize=(20, 5))
ax1 = fig1.add_subplot(1,2,1)
ax1.set_title('Original image')
ax1.imshow(image)
ax2 = fig1.add_subplot(1,2,2)
ax2.set_title('Histogram of original image')
ax2.set_xlabel('Bins')
ax2.set_xlim([0, 255])
ax2.set_ylabel('# of pixels')
ax2.plot(histR, color='r')
ax2.plot(histG, color='g')
ax2.plot(histB, color='b')

#Create mask and masking to image
mask = np.zeros(image.shape[0:2], dtype='uint8')
cv2.rectangle(mask, (15,15), (130, 100), 255, -1)
masked = cv2.bitwise_and(image, image, mask = mask)
chans2 = cv2.split(masked)
mhistR = cv2.calcHist([chans2[0]], [0], mask, [256], [0, 256])
mhistG = cv2.calcHist([chans2[1]], [0], mask, [256], [0, 256])
mhistB = cv2.calcHist([chans2[2]], [0], mask, [256], [0, 256])

#Plot and show masked
fig2 = plt.figure(figsize=(20, 5))
ax1 = fig2.add_subplot(1,3,1)
ax1.set_title('Mask')
ax1.imshow(mask, cmap='gray')
ax2 = fig2.add_subplot(1,3,2)
ax2.set_title('Masked image')
ax2.imshow(masked)
ax3 = fig2.add_subplot(1,3,3)
ax3.set_title('Histogram of masked image')
ax3.set_xlabel('Bins')
ax3.set_xlim([0, 255])
ax3.set_ylabel('# of pixels')
ax3.plot(mhistR, color='r')
ax3.plot(mhistG, color='g')
ax3.plot(mhistB, color='b')
```
